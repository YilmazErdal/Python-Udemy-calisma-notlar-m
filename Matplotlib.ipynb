```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,8)
```

```python
x
```

```python
y = np.arange(1,15,2)
```

```python
y
```

```python
plt.plot(x,y,"green")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"red")
plt.subplot(2,2,2)
plt.plot(y,x,"blue")
plt.subplot(2,2,3)
plt.plot(y,y*2,"yellow")
plt.subplot(2,2,4)
plt.plot(x,x**2,"green")
plt.show()
```

```python
fig = plt.figure()  # figur olusturmak için.

axes1 = fig.add_axes([0.2,0.1,0.8,0.9]) # 1.figürümüzü oluştu.sola 0.2 alta 0.1 uzaklığı.0.8 x uzunluğu 0.9 y uzunluğu
axes2 = fig.add_axes([0.3,0.5,0.3,0.4]) # 2.figürümüz.aynı mantık

axes1.plot(x,y)                    # figürlerimiz içine grafik çizme işi
axes1.set_xlabel("outer x")        # dıştaki figürün grafiği
axes1.set_ylabel("outer y")
axes1.set_title("Outer Grafh")

axes2.plot(y,x)                    # figürlerimiz içine grafik çizme işi
axes2.set_xlabel("inner x")        # içteki figürün grafiği
axes2.set_ylabel("inner y")
axes2.set_title("Inner Grafh")

plt.show()

```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)  # 4 tane axes oluşturmuş oladuk. 2 satır ve 2 sütun 4 axes.
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)  # 4 tane axes oluşturmuş oladuk. 2 satır ve 2 sütun 4 axes.

plt.tight_layout()   #  figürü biraz daha ayrık yapmak için kullandığımız fonc.
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].plot(x,y)
axes[0].set_title("First Axes")  # axesimize başlık verdik
axes[1].plot(y,x)
axes[1].set_title("Second Axes")  # axesimize başlık verdik

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (4,5))    # figürümüzün size(büyüklüğünü) verebiliyoruz.

axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,x*y, color = "blue")
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (7,6))

axes[0].plot(x,y, color = "yellow")
axes[0].set_title("First Axes")
axes[1].plot(y,x**2, color = "red")
axes[1].set_title("Second Axes")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (6,4))   # yeni bir figür oluşturduk. size(6,4)olarak verdik

axes = fig.add_axes([0,0,1,1])    # Figürümüze tüm figürü kaplayacak şekilde([0,0,1,1]) axes ekledik.

axes.plot(x,x**0.5, color = "blue", label = "x karekök")  # grafiklerimizin rengini verdik ve bir isim atadık.
axes.plot(x,x**2, color = "red", label = "x kare")
axes.plot(x,x**3, color = "#d923fd" , label = "x küp")


axes.legend()   # Grafiklerin isimlerinin görünmesi için bu fonction kullanılır.
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, color = "blue", lw = 3,ls = "--", marker = "o",markersize = 10 , markerfacecolor = "black" , 
         markeredgecolor = "red", markeredgewidth = 3)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, color = "blue", lw = 3,ls = "--", marker = "o",markersize = 10 , markerfacecolor = "black" , 
         markeredgecolor = "red", markeredgewidth = 3)  # grafiğimize özellikler katabiliyoruz.

axes.set_xlim(0,20)  # x ekseninde 20(girdiğimiz sayı) ye kadar gösterir
axes.set_ylim(0,100)  # y ekseninde 100(girdiğimiz sayı) e kadar gösterir
```
